```python
import pandas as pd
import zipfile

# Define the file paths
credits_file = 'credits.csv.zip'
movies_file = 'movies_metadata.csv.zip'
ratings_file = 'ratings_small.csv.zip'

# Extract the files from the zip archives
with zipfile.ZipFile(credits_file, 'r') as zip_ref:
    zip_ref.extractall()

with zipfile.ZipFile(movies_file, 'r') as zip_ref:
    zip_ref.extractall()

with zipfile.ZipFile(ratings_file, 'r') as zip_ref:
    zip_ref.extractall()

# Load the CSV files into DataFrames
credits_df = pd.read_csv('credits.csv')
movies_df = pd.read_csv('movies_metadata.csv')
ratings_df = pd.read_csv('ratings_small.csv')

# Check for NaN values in each DataFrame
credits_nan_values = credits_df.isna().sum()
movies_nan_values = movies_df.isna().sum()
ratings_nan_values = ratings_df.isna().sum()

# Print the NaN values count for each DataFrame
print("NaN values count in credits.csv:")
print(credits_nan_values)

print("\nNaN values count in movies_metadata.csv:")
print(movies_nan_values)

print("\nNaN values count in ratings_small.csv:")
print(ratings_nan_values)
```

```python
import pandas as pd

# Read the movies_metadata.csv file
movies_df = pd.read_csv('movies_metadata.csv')

# Create a copy of the DataFrame
new_movies_df = movies_df.copy()

# Impute the NaN values in the 'belongs_to_collection' column with "No Collection"
new_movies_df['belongs_to_collection'].fillna("No Collection", inplace=True)

new_movies_df['homepage'].fillna("No Homepage", inplace=True)

# Delete all the rows with NaN values in the 'imdb_id' column
new_movies_df.dropna(subset=['imdb_id'], inplace=True)

new_movies_df['original_language'].fillna("en", inplace=True)

new_movies_df['overview'].fillna("No Overview", inplace=True)

movies_df['popularity'] = pd.to_numeric(movies_df['popularity'], errors='coerce')
average_popularity = movies_df['popularity'].mean()  
new_movies_df['popularity'].fillna(average_popularity, inplace=True)

# Drop the 'poster_path' column from the DataFrame
new_movies_df.drop('poster_path', axis=1, inplace=True)

movies_df['production_companies'] = movies_df['production_companies'].astype(str)
new_movies_df['production_companies'].fillna("Unknown Companies", inplace=True)

movies_df['production_countries'] = movies_df['production_companies'].astype(str)
new_movies_df['production_countries'].fillna("Unknown Countries", inplace=True)

movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
median_release_date = movies_df['release_date'].median()
new_movies_df['release_date'].fillna(median_release_date, inplace=True)


average_revenue = movies_df['revenue'].mean()  
new_movies_df['revenue'].fillna(average_revenue, inplace=True)

average_runtime = movies_df['runtime'].mean()  
new_movies_df['runtime'].fillna(average_runtime, inplace=True)

new_movies_df['spoken_languages'].fillna("Spoken Languages Unknown", inplace=True)

new_movies_df['status'].fillna("Released", inplace=True)

new_movies_df['tagline'].fillna("No tagline", inplace=True)

new_movies_df['title'].fillna("Unknown title", inplace=True)

new_movies_df['video'].fillna(False, inplace=True)

average_vote_average = movies_df['vote_average'].mean()  
new_movies_df['vote_average'].fillna(average_vote_average, inplace=True)

average_vote_count = movies_df['vote_count'].mean()  
new_movies_df['vote_count'].fillna(average_vote_count, inplace=True)

# Save the new DataFrame to a new file
new_movies_df.to_csv('movies_metadata_imputed.csv', index=False)
```
